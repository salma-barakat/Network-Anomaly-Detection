```python
def kmeans(data,no_clusters=3,max_iteration=100,init_seeds=None):
  t=0
  repeat = 1
  rows=data.shape[0]
  columns=data.shape[1]
  nu_clusters=no_clusters ## number of clusters
  max_iter=max_iteration  ## maximum number of iterations
  distance=np.zeros((rows,nu_clusters)) ## a distance matrix with dimensions data size x number of clusters
  clusters=np.zeros(rows) ## an array to store the equivalent cluster number of each node


  ## if there is no initial centoids then pick random centroids 
  ## else assign them the old centroids to be the first centroids of the cluster
  if (init_seeds is None):
    old_centroids = data.sample(n=nu_clusters)
  else:
    old_centroids=init_seeds

  unique = old_centroids.drop_duplicates().shape[0]
  boolean = old_centroids.duplicated().any()
  ## check that all the centroids are unique and there are no duplicates
  while(boolean):
    old_centroids = old_centroids.drop_duplicates()
    difference = nu_clusters - unique
    old_centroids2 = data.sample(n=difference)
    old_centroids = pd.concat([old_centroids,old_centroids2],ignore_index=True)
    unique = old_centroids.drop_duplicates().shape[0]
    boolean = old_centroids.duplicated().any()
  new_centroids=np.zeros(old_centroids.shape)
  while (t<max_iteration):
    t=t+1
    #old_centroids = new_centroids
    # measure the distance of each row to every cluster
    for i in range (nu_clusters):
      distance[:,i]=np.linalg.norm(data-old_centroids.iloc[i,:],axis=1)
    # assign data to closest centroid
    clusters=np.argmin(distance, axis=1)
    # update centroid
    for i in range (nu_clusters):
      old_centroids.iloc[i,:]=np.mean(data[clusters==i],axis=0)
      # new_centroids[i,:]=np.mean(data[clusters==i],axis=0)
    # if new_centroids == old_centroids:
    #   break
    # old_centroids = deepcopy(new_centroids)
  return old_centroids,clusters
```

```python
def evaluate(clusters,new_centroids,testx,testy): 
    ## cluster the testing set
    rows=testx.shape[0]
    nu_clusters = len(np.unique(clusters))
    distance=np.zeros((rows,nu_clusters))
    test_clusters=np.zeros(rows)
    new_centroids = new_centroids.to_numpy()
    ## calculate the distance between the test data and the centroids from the kmeans
    for i in range (nu_clusters):
        distance[:,i]=np.linalg.norm(testx-new_centroids[i,:],axis=1)
    ## assign data to closest centroid
    test_clusters=np.argmin(distance, axis=1)
    print("Unique test cluster:")
    print(np.unique(test_clusters))
    print(test_clusters)


    d=testy  ##truth array
    t=test_clusters ##output array

    t_unique,t_count = np.unique(t,return_counts=True)
    d_unique,d_count = np.unique(d,return_counts=True)
    

    clus_row = nu_clusters
    clus_col = np.max(testy)+1
    print("@K = ",clus_row)
    count = np.zeros([clus_row,clus_col])
    ## count the number of different data in the clusters
    for i in range(len(t)):
       count[t[i]][d[i]] = count[t[i]][d[i]] +1    ## rows: output clusters (k), columns: testing clusters (40)
    row_sum= np.sum(count,axis=1) ## the sum of data in each cluster
    column_sum= np.sum(count,axis=0) ## the sum of data in different group of the test data
    total_row_sum = np.sum(row_sum) ## the total number of test data

    htc = np.zeros([1,clus_row])
    print("The count array size:", count.shape)


    ##Conditional Entropy
    print("------------------------------------------------------------------------------")
    print("Conditional Entropy")
    for i in range(clus_row):
      for j in range(clus_col):
        ## calculate H(T/Ci)
        if (count[i][j]):
          htc[0][i] = htc[0][i] - ( count[i][j]/row_sum[i] * math.log(count[i][j]/row_sum[i],10) )
    print(htc)
    htc_total = 0
    for i in range(clus_row):
      ## calculate the total entropy
      htc_total = htc_total + ( row_sum[i]/total_row_sum * htc[0][i] )
    print("Condition Entropy = ", htc_total)



    ##Precision and Recall
    print("------------------------------------------------------------------------------")
    print("Precision and Recall")
    max_element = np.max(count,axis=1) ## choose the majority element
    index_sum = np.zeros([1,len(max_element)])
    ## find the index of the major element in each cluster
    for i in range(len(max_element)):
      temp = np.argwhere(count == max_element[i])
      for j in temp:
        if (j[0] == i):
          index_sum[0][i] = column_sum[j[1]]

    prec = np.zeros([1,len(max_element)])
    total_prec=0 
    total_rec=0 
    rec = np.zeros([1,len(max_element)])

    for i in range(len(max_element)):
      ## calculate the precision of the cluster and the total precision
      prec[0][i] = max_element[i]/row_sum[i]
      prec[np.isnan(prec)] = 0
      total_prec = total_prec + (row_sum[i]/rows)*prec[0][i]
      ## calculate the recall of the cluster and the total recall
      rec[0][i] = max_element[i]/index_sum[0][i]      
      total_rec = total_rec + (row_sum[i]/rows)*rec[0][i]

    print("Precision: ")
    print(prec)
    print("Total Precision = ",total_prec)
  
    print("Recall:")
    print(rec)
    print("Total Recall = ",total_rec)
    

    ##F-score
    print("------------------------------------------------------------------------------")
    print("F-score")
    f_cluster = 2 * np.multiply ( np.multiply( prec , rec) , 1/(prec + rec))
    print("F-cluster: ")
    f_cluster[np.isnan(f_cluster)] = 0
    print(f_cluster)
    f_measure = np.sum(f_cluster,axis=1) * 1/clus_row
    print("F-score: ")
    f_measure[np.isnan(f_measure)] = 0
    print(f_measure)  
```

```python
def Similarity_knn(data,k,neigh):
    ## find the nearest k neighbours
    knn=NearestNeighbors(n_neighbors=neigh)
    knn.fit(data)
    #Compute the (weighted) graph of k-Neighbors for points in data.
    sim_matrix = knn.kneighbors_graph(data, mode='connectivity').toarray() 
    # compute delta matrix( sum of internal & external weights)
    delta_matrix=np.diag(np.sum(sim_matrix,axis=1))
    # get normalized asymmetric laplacian matrix
    l=delta_matrix-sim_matrix
    la=np.dot(np.linalg.inv(delta_matrix),l)
    #get eigenvalues& eigenvectors
    eigenvalues,eigenvectors=np.linalg.eig(la)
    # sort eigenvalues&eigenvectors and take number of eigenvectors equal to number of clusters     
    idx=eigenvalues.argsort()[:k]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:,idx]
    # normalize eigenvectors
    rows_norm=np.linalg.norm(eigenvectors,axis=1) 
    vectors=np.real(np.transpose((np.transpose(eigenvectors)/rows_norm)))
    return eigenvalues,vectors
```

###There is 2 major parameters for DBSCAN clustering.
1-Epsilon
This is the radius of the circle that we must draw around our pointing focus.

2-Minimum number of points
This is the number of points that we want in the neighbourhood of our point in focus (within the circle).

**DBSCAN** checks for clusters by checking the ε-neighborhood of every point in the database. If the ε-neighborhood of a point p contains more than MinPts, a new cluster with p as a core element is produced. DBSCAN iteratively assemble precisely density-reachable objects from these essential element, which can include the merge of a few density-reachable clusters. The process eliminates when no new point can be added to any cluster.

```python
#split_train, split_test, splity_train, splity_test = train_test_split(xtrain_total, ytrain_total, test_size=0.99975, random_state=42,stratify=ytrain_total)

#function to get the neighbours of a point
#points lying inside the circle of radius = epsilon
def insideCircle(D, p, epsilon):
  neighbors = []
  for pn in range(len(D)):
    # If the distance is within the radius epsilon, add it to the neighbors list.
    if np.linalg.norm(D[p] - D[pn]) < epsilon:
      neighbors.append(pn)          
  return neighbors

#to get neighbours of neighbours:
def expandCluster(D, labels, p, neighPoints, Cluster_id, epsilon, min_points):
  labels[p] = Cluster_id 
  i = 0
  while i < len(neighPoints):    
      # Get the next point.        
      pn = neighPoints[i]
      if labels[pn] == -1: #noise
          labels[pn] = Cluster_id
      elif labels[pn] == 0: #not visited before
          labels[pn] = Cluster_id
          # Find the neighbors of each pn:
          PnNeighborPts = insideCircle(D, pn, epsilon)
          if len(PnNeighborPts) >= min_points: #pn is not noise
              neighPoints = neighPoints + PnNeighborPts
      i += 1        
  
    # We've finished growing cluster C!

D = split_train.to_numpy()
labels = [0]*len(D)  
epsilon = 50
min_points = 35
Cluster_id = 0 #default
for p in range(len(D)):
  if(labels[p]==0):
    neighPoints = insideCircle(D, p, epsilon)
    if len(neighPoints) < min_points:
      labels[p] = -1
    else:
      Cluster_id += 1 
      expandCluster(D, labels, p, neighPoints, Cluster_id, epsilon, min_points)
print(labels)
```
